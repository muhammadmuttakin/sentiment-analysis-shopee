```python
import pandas as pd  # Untuk mengelola data dan menyimpan ke CSV
from google_play_scraper import reviews, Sort  # Untuk scraping ulasan dari Google Play
```

```python
# Scraping ulasan aplikasi "RRI Play Go"
app_id = "com.shopee.id"
result, _ = reviews(
    app_id,
    lang='id',  # Bahasa Indonesia
    country='id',  # Negara Indonesia
    count=12000,  # Ubah jumlah ulasan jika terlalu besar
    sort=Sort.NEWEST  # Gunakan enum Sort.NEWEST
)

# Konversi hasil scraping ke DataFrame
reviews_df = pd.DataFrame(result)

# Simpan ke file CSV
reviews_df.to_csv("data_review.csv", index=False)

print("Scraping selesai! Data disimpan di data_review.csv")
```
